```python
import numpy as np
```

```python
a=np.arange(0,20,2)
```

```python
print(a)
```

```python
b=a*2
print(b)
```

```python
c=a**2
```

```python
print(c)
```

```python
d=np.sin(a)
```

```python
print(d)
```

```python
e=np.cos(a)
print(e)
```

```python
data = np.array([a,b,c,d,e])

data.shape
```

```python
data = np.transpose(data)
print(data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
df.columns=['a','b','c','d','e']
```

```python
df.to_csv('demo.csv')
```
